```python
import pandas as pd
import numpy as np
```

```python
#import data BTC
df_BTC = pd.read_csv('BTC_USD Bitfinex Historical Data.csv',index_col='Date',parse_dates=True)
```

```python
df_BTC
```

```python
#Xóa dấu , và chuyển về float
df_BTC['Price']=df_BTC['Price'].replace(',','',regex=True).astype(float)
df_BTC['Open']=df_BTC['Open'].astype(str).str.replace(',','').astype(float)
df_BTC['High']=df_BTC['High'].astype(str).str.replace(',','').astype(float)
df_BTC['Low']=df_BTC['Low'].astype(str).str.replace(',','').astype(float)
```

```python
#Thực hiện tách chuỗi KL thành 1 chuỗi số float và 1 ký tự cuối sau đó lấy số float đó nhân với 1000 
#nếu ký tự cuối là K, nếu là M thì nhân với 1000000, nếu không có K hay M thì trả về số float đó
def convert_value(value):
    if isinstance(value, str):
        if value.endswith('K'):
            return float(value[:-1]) * 1000
        elif value.endswith('M'):
            return float(value[:-1]) * 1000000
        elif value.endswith('B'):
            return float(value[:-1]) * 1000000000
        else:
            return float(value)
    else:
        return value
```

```python
df_BTC['Vol.'] = df_BTC['Vol.'].apply(convert_value)
```
